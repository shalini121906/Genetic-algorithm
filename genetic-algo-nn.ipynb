```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=40)
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from sklearn.metrics import accuracy_score
```

```python
nn_in=X_train.shape[1]
nn_out = 1
nn_hidden=8

def network(nn_in,nn_out,nn_hidden):
    model=Sequential()
    model.add(Dense(nn_hidden,input_shape=(nn_in,),activation='relu',name='lay1'))
    model.add(Dense(nn_out,activation='sigmoid',name='lay2'))
    model.compile(loss='mse',optimizer='adam',metrics=['accuracy'])
    print(model.summary())
    return model
```

```python
def run(X_train,y_train,X_test,y_test,nn_in,nn_out,nn_hidden,accuracies,weights):
    model=network(nn_in,nn_out,nn_hidden)
    model.fit(X_train,y_train)
    accuracy=model.evaluate(X_test,y_test)
    weight=model.get_weights()[0]
    accuracies.append(accuracy[1])
    weights.append(weight)
```

```python
accuracies=[]
weights=[]
for i in range(10):
    model1=run(X_train,y_train, X_test, y_test,nn_in,nn_out,nn_hidden,accuracies,weights)
```

```python
def selectionfunc(accuracies,weights):
    res = {}
    for key in accuracies:
        for value in weights:
            res[key] = value
            weights.remove(value)
            break
    myKeys = list(res.keys())
    myKeys.sort(reverse=True)
    sorted_dict = {i: res[i] for i in myKeys}
    return sorted_dict
```

```python
def fitnessfunc(child_weights):
    acc=[]
    model1=Sequential()
    model1.add(Dense(nn_hidden,input_shape=(nn_in,),activation='relu'))
    model1.add(Dense(nn_out,activation='sigmoid',))
    model1.compile(loss='mse',optimizer='adam',metrics=['accuracy'])
    for i in range(len(child_weights)):
        model1.layers[0].set_weights([child_weights[i], np.zeros(model1.layers[0].get_weights()[1].shape)])
        model1.fit(X_train,y_train)
        accuracy=model1.evaluate(X_test,y_test)
        acc.append(accuracy[1])
    return acc
```

```python
def crossover1(parent1, parent2):
    # Get the dimensions of the parents
    # Get the dimensions of the parents
    n_rows, n_cols = parent1.shape

    # Initialize the child chromosome
    child = np.zeros((n_rows, n_cols))

    # Iterate over each gene in the parents' chromosomes and select it for the child
    for i in range(n_rows):
        for j in range(n_cols):
            if np.random.choice([True, False]):
                child[i, j] = parent1[i, j]
            else:
                child[i, j] = parent2[i, j]
    
    mutationfunc(child)
      
    return child
```

```python
def mutationfunc(child_weights):
    selection = random.randint(0, len(child_weights)-1)
    mut = random.uniform(0, 1)
    if mut <= .05:
        child_weights[selection] *= random.randint(2, 5)
    else:
        pass
```

```python
genr=0
child_weights=[]
o=[]
p=[]
def generationfunction(accuracies,weights,genr,o,p,child_weights):
    items = Counter(accuracies).keys()
    if(len(items)==1):
            
            print("Cant run the model anymore as all the accuracies obtained are the same")
            print("  ")
            exitloop()
    else:
        if(genr!=10):
            genr+=1
            print("Generation :",genr)
            sorted_dict=selectionfunc(accuracies,weights)
            accdic = list(sorted_dict.items())
            for j in range(2):
                o.append(accdic[j][1])
            for j in range(2,len(accdic)):
                p.append(accdic[j][1])
            for j in range(10):
                p1=random.choice(o)
                p2=random.choice(p)
                child_weights.append(crossover1(p1,p2))
            child_w=child_weights.copy()
            o.clear()
            p.clear()
            child_weights.clear()
            acc=fitnessfunc(child_w)
            print("The Accuracies of this generation: ")
            print(acc)

            generationfunction(acc,child_w,genr,o,p,child_weights)

   
```
